```python
import os
import h5py
import numpy as np
```

```python
HOME_PATH = os.path.expanduser(f'~')
USERNAME = os.path.split(HOME_PATH)[1]
PATH_DATASETS = 'datasets/Thyroid_Ultrasound_Cine-clip/dataset.hdf5'
FULL_PATH_FOR_HDF5_FILE = os.path.join(HOME_PATH, PATH_DATASETS)
# print(FULL_PATH_FOR_HDF5_FILE)
```

```python
f = h5py.File(FULL_PATH_FOR_HDF5_FILE, 'r')

for key in f.keys():
    print(key) #Names of the root level object names in HDF5 file - can be groups or datasets.
    print(type(f[key])) # get the object type: usually group or dataset
```

```python
annot_id_ = f.get('annot_id')
print(annot_id_)
print(annot_id_.shape)

frame_num_ = f.get('frame_num')
print(frame_num_)
print(frame_num_.shape)

image_ = f.get('image')
print(image_)
print(image_.shape)

mask_ = f.get('mask')
print(mask_)
print(mask_.shape)
```
